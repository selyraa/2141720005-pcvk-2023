```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
img_4 = cv.imread('./KTP.jpg')
img_5 = cv.cvtColor(img_4, cv.COLOR_BGR2RGB)
cv2_imshow(img_4)
plt.imshow(img_5)
```

```python
img_4s = np.copy(img_4)
#tutup bagian agama
for y in range (190,260):
  for x in range (280,300):
    img_4s[x,y] = [0,255,0]

#tutup bagian status kawin
img_4ss = np.copy(img_4s)
for y in range (190,270):
  for x in range (305,325):
    img_4ss[x,y] = [255,0,0]
final_frame = cv.hconcat((img_4, img_4ss))
cv2_imshow(final_frame)
```
